```python
from bs4 import BeautifulSoup
import requests

url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup
```

```python
soup = soup.text
```

```python
soup = BeautifulSoup(soup, "lxml")
soup
```

```python
table = soup.table
table
```

```python
tags = table.find_all('tr')
tags
```

```python
data = []
for tag in tags:
  x = tag.text
  x = x.split('\n')
  data.append(x[1:-1])

print(data)
```

```python
final_data = []
for j in data:
  if j[0] != '':
    final_data.append(j)

print(final_data)
```

```python
import csv
file = open('covid_data1.csv','w')
x = csv.writer(file)
for i in final_data:
  x.writerow(i)
```

```python
import pandas as pd
df = pd.read_csv('covid_data1.csv')
df
```

```python
deaths = []
countries = list(df['Country,Other'][0:10])
TotalDeaths = list(df['TotalCases'][0:10])
for i in TotalDeaths:
  v = int(i.replace(',',''))
  deaths.append(v)
```
